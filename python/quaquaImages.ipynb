```python
def get_album_image(artist, album):
    try:
        search = f'{artist} {album} Album Wikipedia'
        url = 'https://www.google.com/search'

        headers = {
            'Accept' : '*/*',
            'Accept-Language': 'en-US,en;q=0.5',
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.82',
        }
        parameters = {'q': search}

        content = requests.get(url, headers = headers, params = parameters).text
        soup = BeautifulSoup(content, 'html.parser')

        search = soup.find(id = 'search')
        first_link = search.find('a')
        link = first_link['href']

        wiki = requests.get(link)

        # Parse the HTML content
        soup = BeautifulSoup(wiki.content, 'html.parser')

        # Find the infobox and the first image inside it
        infobox = soup.find('table', class_='infobox')
        image_tag = infobox.find('img')

        # Extract the image URL
        if image_tag:
            image_url = image_tag['src']
            if image_url.startswith("//"):
                image_url = "https:" + image_url
                print(image_url)
            return image_url
        else:
            return ''
    except Exception as e:
        print(f"An error occurred: {e}")  # Optionally log the error
        return ''  # Return empty string on any exception

```

```python
import wikipedia

def get_wikipedia_page(artist, album):
    search = f'{artist} {album} Album Wikipedia'

    try:
        page = wikipedia.page(search)

        try:
            # Use the page.url to get the Wikipedia page
            wiki = requests.get(page.url)

            # Parse the HTML content
            soup = BeautifulSoup(wiki.content, 'html.parser')

            # Find the infobox and the first image inside it
            infobox = soup.find('table', class_='infobox')
            if infobox:
                image_tag = infobox.find('img')

                # Extract the image URL
                if image_tag:
                    image_url = image_tag['src']
                    if image_url.startswith("//"):
                        image_url = "https:" + image_url
                        print(image_url)
                    return image_url
            return ''  # Return empty string if no image is found
        except Exception as e:
            print(f"An error occurred while fetching the Wikipedia page: {e}")
            return ''  # Return empty string on any exception

    except wikipedia.exceptions.DisambiguationError as e:
        print(f"Disambiguation page found. Options: {e.options}")
        return ''
    except wikipedia.exceptions.PageError:
        print(f"No page found")
        return ''
        

```
