```python
df = pd.read_csv('top_artists.csv')
df.head()
```

```python

def get_album_playcount(artist, album):
    print(album)

    # API endpoint
    url = f'http://ws.audioscrobbler.com/2.0/?method=album.getinfo&api_key={API_KEY}&artist={artist}&album={album}&format=json'

    # Make the request
    response = requests.get(url)

    if response.status_code == 200:
        album_info = response.json()
        
        # Extract the playcount
        playcount = album_info.get('album', {}).get('playcount', 'N/A')
        return playcount
    else:
        return 0
```

```python
df.loc[:, 'lastfm_playcount'] = df.apply(lambda row : get_album_playcount(row['Artist'],row['Album']), axis=1)
```
